# If Condition

```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df['price_label'] = df.apply(price_label, axis=1)
```

```python
df['price_label'].value_counts()
```

```python
df['prices'].max()
```

# If-Statements with the loc() Function

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

# 2. Task

```python
# printing the frequency of the “orders_day_of_week” 
df_orders_products_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
result = []

for value in df_orders_products_merged["orders_day_of_week"]:
  if value <= 1:
    result.append("Busiest days")
  elif value >= 3 and value <=4:
    result.append("Slowest days")  
  else:
    result.append("Regularly busy")
```

```python
# create a new column and set it equal to result
df_orders_products_merged['Busiest days'] = result
```

# 3. Task

```python
# print the frequency of this new column 
df_orders_products_merged['Busiest days'].value_counts(dropna = False)
```
