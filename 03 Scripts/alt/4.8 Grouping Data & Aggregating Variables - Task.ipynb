# 4. Task

```python
# creat new column max_order
ords_prods_merge2['max_order'] = ords_prods_merge2.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge2.head()
```

```python
# not to assign any options regarding the maximum number of rows to display
pd.options.display.max_rows = None
```

```python
# check these values by scrolling through the user_ID column and matching the max order value to the number of times the user ID appears within the dataframe
ords_prods_merge2.head(100)
```

```python
# create a flag that assigns a loyalty label to a user ID based on its corresponding max order value
ords_prods_merge2.loc[ords_prods_merge2['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge2.loc[(ords_prods_merge2['max_order'] <= 40) & (ords_prods_merge2['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge2.loc[ords_prods_merge2['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge2['loyalty_flag'].value_counts(dropna = False)
```

# 5. Task

```python
# showing the difference between the spending habits 
ords_prods_merge2.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

What youre trying to determine is whether the prices of products purchased by loyal customers differ from those purchased by regular or new customers.
Answer: The mean price of new customers is higher than the mean price of loyal and regular customers.

# 7. Task

```python
# creat new column mean_order_frequency
ords_prods_merge2_clean['median_order_frequency'] = ords_prods_merge2_clean.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge2_clean.head()
```

```python
# create a flag 
ords_prods_merge2_clean.loc[ords_prods_merge2_clean['median_order_frequency'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge2_clean.loc[(ords_prods_merge2_clean['median_order_frequency'] <= 20) & (ords_prods_merge2_clean['median_order_frequency'] > 10), 'order_frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge2_clean.loc[ords_prods_merge2_clean['median_order_frequency'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge2_clean['order_frequency_flag'].value_counts(dropna = False)
```
