# Checking Mixed-type Columns

```python
for col in df2.columns.tolist():
  weird = (df2[[col]].applymap(type) != df2[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df2[weird]) > 0:
    print (col)
```

```python
#changing types of these columns
df2['heatingType'] = df2['heatingType'].astype('str')
df2['telekomTvOffer'] = df2['telekomTvOffer'].astype('str')
df2['firingTypes'] = df2['firingTypes'].astype('str')
df2['condition'] = df2['condition'].astype('str')
df2['interiorQual'] = df2['interiorQual'].astype('str')
df2['petsAllowed'] = df2['petsAllowed'].astype('str')
df2['typeOfFlat'] = df2['typeOfFlat'].astype('str')
df2['energyEfficiencyClass'] = df2['energyEfficiencyClass'].astype('str')
```
